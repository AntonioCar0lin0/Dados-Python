```python
import pandas as pd
```

```python
df = pd.read_csv('/content/dataset_sintetico.csv')
```

```python
df
```

```python
horas = df[['Horas de Estudo']]
nota = df['Nota na Prova']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(horas, nota, c='red')
plt.title('GrÃ¡fico de dispersÃ£o original')
plt.xlabel('Horas de estudo')
plt.ylabel('Nota na prova')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Treinando o modelo
modelo = LinearRegression()
modelo.fit(horas, nota)
```

```python
#Relaizando a previsÃ£o das notas com base nas horas de estudo
nota_pred = modelo.predict(horas)
```

```python
import matplotlib.pyplot as plt
```

```python
#VisualizaÃ§Ã£o dos grÃ¡ficos
plt.scatter(horas, nota, label='Dados originais')
plt.plot(horas, nota_pred, label='Linha de regressÃ£o', c = 'red')
plt.title('Horas de estudo x Notas na prova')
plt.xlabel('Horas de estudo')
plt.ylabel('Nota na prova')
plt.legend()
plt.show()
```

InterpretaÃ§Ã£o do grÃ¡fico: Com a linha de regressÃ£o Ã© possÃ­vel prever a nota aproximada do estudante com base no nÃºmero de horas de estudo.
Por exemplo, se um estudante estudar por 2 horas, ele irÃ¡ tirar 10 na prova.
Se tirar 6, tirarÃ¡ 30 aproximadamente.

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(nota, nota_pred)
r2 = r2_score(nota, nota_pred)
print(f'Erro quadrÃ¡tico mÃ©dio: {mse:.2f}')
print(f'Coeficiente de determinaÃ§Ã£o: {r2*100:.2f}%')
```

**Erro quadrÃ¡tico mÃ©dio**: Ã‰ uma medida que calcula a mÃ©dia dos quadrados da diferenÃ§a entre o quadrado dos valores reais pelos valores previstos pelo modelo.
Quanto menor o erro quadrÃ¡tico mÃ©dio, melhor o desempenho do modelo.

**Coeficiente de determinaÃ§Ã£o:** Ã‰ uma medida que compara o desempenho do modelo com um modelo simples a mÃ©dia dos valores reais como previsÃ£o. Em outras palavras, explica o quanto a variaÃ§Ã£o dos dados o modelo conseguiu explicar
