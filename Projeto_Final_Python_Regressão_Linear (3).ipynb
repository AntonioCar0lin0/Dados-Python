```python
import pandas as pd
```

```python
df = pd.read_csv('/content/dataset_sintetico.csv')
```

```python
df
```

```python
horas = df[['Horas de Estudo']]
nota = df['Nota na Prova']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(horas, nota, c='red')
plt.title('Gráfico de dispersão original')
plt.xlabel('Horas de estudo')
plt.ylabel('Nota na prova')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Treinando o modelo
modelo = LinearRegression()
modelo.fit(horas, nota)
```

```python
#Relaizando a previsão das notas com base nas horas de estudo
nota_pred = modelo.predict(horas)
```

```python
import matplotlib.pyplot as plt
```

```python
#Visualização dos gráficos
plt.scatter(horas, nota, label='Dados originais')
plt.plot(horas, nota_pred, label='Linha de regressão', c = 'red')
plt.title('Horas de estudo x Notas na prova')
plt.xlabel('Horas de estudo')
plt.ylabel('Nota na prova')
plt.legend()
plt.show()
```

Interpretação do gráfico: Com a linha de regressão é possível prever a nota aproximada do estudante com base no número de horas de estudo.
Por exemplo, se um estudante estudar por 2 horas, ele irá tirar 10 na prova.
Se tirar 6, tirará 30 aproximadamente.

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(nota, nota_pred)
r2 = r2_score(nota, nota_pred)
print(f'Erro quadrático médio: {mse:.2f}')
print(f'Coeficiente de determinação: {r2*100:.2f}%')
```

**Erro quadrático médio**: É uma medida que calcula a média dos quadrados da diferença entre o quadrado dos valores reais pelos valores previstos pelo modelo.
Quanto menor o erro quadrático médio, melhor o desempenho do modelo.

**Coeficiente de determinação:** É uma medida que compara o desempenho do modelo com um modelo simples a média dos valores reais como previsão. Em outras palavras, explica o quanto a variação dos dados o modelo conseguiu explicar
