#Análise exploratória de dados



*   Carregue os dados usando Pandas





````python
#Importando pandas para visualização dos dados
import pandas as pd
````

````python
#Lendo o arquivo .csv usando pandas
df = pd.read_csv('/content/iris.csv')
df
````

````python
#Atribuindo variáveis às colunas
sepala_comprimento = df['sepal.length']
sepala_largura = df['sepal.width']
petala_comprimento = df['petal.length']
petala_largura = df['petal.width']
especies = df['variety']
````



*   Calcule estatísticas descritivas (média, mediana, desvio padrão) para cada atributo. Objetivo: Familiarizar-se com a manipulação de dados com Pandas e a visualização com Matplotlib.




````python
import numpy as np
````

````python
#Médias gerais
media_sepala_comprimento = np.mean(sepala_comprimento)
print(f'A média de comprimento das sépalas no conjunto de dados é: {media_sepala_comprimento:.2f}\n')

media_sepala_largura = np.mean(sepala_largura)
print(f'A média de largura das sépalas no conjunto de dados é: {media_sepala_largura:.2f}\n')

media_petala_comprimento = np.mean(petala_comprimento)
print(f'A média de comprimento das pétalas no conjunto de dados é: {media_petala_comprimento:.2f}\n')

media_petala_largura = np.mean(petala_largura)
print(f'A média de largura das pétalas no conjunto de dados é: {media_petala_largura:.2f}')
````

````python
#Dados do comprimento da sépala por espécie
media_comprimento_sepala_especie = df.groupby('variety')['sepal.length'].mean() #usar o nome das colunas porque o groupby espera isso
print('Média de comprimento da sépala por espécie')
display(media_comprimento_sepala_especie)

print('\n')

desvio_comprimento_sepala_especie = df.groupby('variety')['sepal.length'].std()
print('Desvio padrão do comprimento da sépala por espécie')
display(desvio_comprimento_sepala_especie)

print('\n')

mediana_comprimento_sepala_especie = df.groupby('variety')['sepal.length'].median()
print('Mediana do comprimento da sépala por espécie')
display(mediana_comprimento_sepala_especie)
````

````python
#Dados da largura da sépala por espécie
media_largura_sepala_especie = df.groupby('variety')['sepal.width'].mean()
print('Média de largura da sépala por espécie')
display(media_largura_sepala_especie)

print('\n')

desvio_largura_sepala_especie = df.groupby('variety')['sepal.width'].std()
print('Desvio padrão da sépala por espécie')
display(desvio_largura_sepala_especie)

print('\n')

mediana_largura_sepala_especie = df.groupby('variety')['sepal.width'].median()
print('Mediana da sépala por espécie')
display(mediana_largura_sepala_especie)
````

````python
#Dados do comprimento da pétala por espécie
media_comprimento_petala_especie = df.groupby('variety')['petal.length'].mean()
print('Média do comprimento da pétala por espécie')
display(media_comprimento_petala_especie)

print('\n')
desvio_comprimento_petala_especie = df.groupby('variety')['petal.length'].std()
print('Desvio padrão do comprimento da pétala por espécie')
display(desvio_comprimento_petala_especie)

print('\n')
mediana_comprimento_petala_especie = df.groupby('variety')['petal.length'].median()
print('Mediana do comprimento da pétala por espécie')
display(mediana_comprimento_petala_especie)
````

````python
#Dados da largura da pétala por espécie
media_largura_petala_especie = df.groupby('variety')['petal.width'].mean()
print('Média da largura da pétala por espécie')
display(media_largura_petala_especie)

print('\n')

desvio_largura_petala_especie = df.groupby('variety')['petal.width'].std()
print('Desvio padrão da largura da pétala por espécie')
display(desvio_largura_petala_especie)

print('\n')

mediana_largura_petala_especie = df.groupby('variety')['petal.width'].median()
print('Mediana da largura da pétala por espécie')
display(mediana_largura_petala_especie)
````

* Utilize Matplotlib para criar gráficos de dispersão, histogramas e boxplots para visualizar as relações entre as variáveis.

````python
import matplotlib.pyplot as plt
````



```
# Histogramas
```



````python
petala_comprimento_especie = df.pivot(columns='variety', values='petal.length')

petala_comprimento_especie.plot.hist(bins=15, alpha=0.5, edgecolor='black')

plt.title('Comprimento da pétala por espécie')
plt.xlabel('Comprimento')
plt.ylabel('Frequência')
plt.legend(title='Espécies')
plt.show()
````

````python
petala_largura_especie = df.pivot(columns='variety', values='petal.width')
petala_largura_especie.plot.hist(bins=15, alpha=0.5, edgecolor='black')

plt.title('Largura da pétala por espécie')
plt.xlabel('Largura')
plt.ylabel('Frequência')
plt.legend(title='Espécies')
plt.show()
````

````python
sepala_comprimento_especie = df.pivot(columns='variety', values='sepal.length')
sepala_comprimento_especie.plot.hist(bins=15, alpha=0.5, edgecolor='black')

plt.title('Comprimento da sépala por espécie')
plt.xlabel('Comprimento')
plt.ylabel('Frequência')
plt.legend(title='Espécies')
plt.show()
````

````python
sepala_largura_especie = df.pivot(columns='variety', values='sepal.width')
sepala_largura_especie.plot.hist(bins=15, alpha=0.5, edgecolor='black')

plt.title('Largura da sépala por espécie')
plt.xlabel('Largura')
plt.ylabel('Frequência')
plt.legend(title='Espécies')
plt.show()
````



```
Gráficos de dispersão
```



````python
cores = {'Setosa': 'blue', 'Versicolor': 'green', 'Virginica': 'red'} #define as cores
df['color'] = df['variety'].map(cores) #coloca as cores segundo os valores da coluna 'variety-espécies'

plt.scatter(df['petal.length'], df['petal.width'], c=df['color'], alpha=0.5, edgecolor='black') # cria o gráfico de dispersão
plt.title('Comprimento x Largura da pétala por espécie')
plt.xlabel('Comprimento da pétala')
plt.ylabel('Largura da pétala')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title=' Setosa - Azul \n Versicolor - Verde \n Virginica - Vermelha')
plt.show()
````

````python
cores = {'Setosa': 'blue', 'Versicolor': 'green', 'Virginica': 'red'}
df['color'] = df['variety'].map(cores)

plt.scatter(df['sepal.length'], df['sepal.width'], c=df['color'], alpha = 0.5, edgecolor='black')
plt.title('Comprimento x largura da sépala por espécie')
plt.xlabel('Comprimento da sépala')
plt.ylabel('Largura da sépala')
plt.legend(title=' Setosa - Azul \n Versicolor - Verde \n Virginica - Vermelha')
plt.show()
````



```
Boxplot's
```

````python
dados = [df[df['variety'] == especie]['petal.length'] for especie in df['variety'].unique()]
plt.boxplot(dados, labels=df['variety'].unique(), patch_artist=True)

plt.title('Boxplot do Comprimento da Pétala por Espécie')
plt.xlabel('Espécie')
plt.ylabel('Comprimento da Pétala')
plt.show()
````
